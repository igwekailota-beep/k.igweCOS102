# Connecting to a Database

```python
import psycopg2

# Database connection parameters
host = "localhost"
port = "5432"
dbname = "cos101_db"
user = "postgres"
password = "cos101"

# Establish the connection
try:
    connection = psycopg2.connect(
        host=host,
        port=port,
        dbname=dbname,
        user=user,
        password=password
    )
    print("Connection to PostgreSQL DB successful")

    # Create a cursor object
    cursor = connection.cursor()

    # Execute a SQL query
    cursor.execute("SELECT version();")
    
    # Fetch and print the result of the query
    db_version = cursor.fetchone()
    print(f"PostgreSQL database version: {db_version}")

    # Close the cursor and connection
    cursor.close()
    connection.close()

except Exception as error:
    print(f"Error connecting to PostgreSQL DB: {error}")
```

# Selecting existing Database

```python
import psycopg2

# Database connection parameters
host = "localhost"
port = "5432"
dbname = "cos101_db"
user = "postgres"
password = "cos101"

# Establish the connection
try:
    connection = psycopg2.connect(
        host=host,
        port=port,
        dbname=dbname,
        user=user,
        password=password
    )
    print("Connection to PostgreSQL DB successful")

    # Create a cursor object
    cursor = connection.cursor()

    # Define the SQL SELECT query
    select_query = "SELECT  FROM company;"

    # Execute the query
    cursor.execute(select_query)

    # Fetch all rows from the executed query
    data = cursor.fetchone()

    print("column", " | ", "column2")
    print("------------------------")
        
    # Iterate through the rows and print them
    for row in data:
        print(f"{row[0]}       |  {row[1]}")

    # Close the cursor and connection
    cursor.close()
    connection.close()

except Exception as error:
    print(f"Error connecting to PostgreSQL DB: {error}")
```

```python
import psycopg2
from psycopg2 import sql  # Fix 1: Import the sql module

# Connection parameters
host = "localhost"
port = "5432"
dbname = "cos101_db"
user = "postgres"
password = "cos101"

# Fix 2: Corrected typo in variable name
new_dbname = "cos102_db"

try:
    connection = psycopg2.connect(
        host=host,
        port=port,
        dbname=dbname,
        user=user,
        password=password
    )
    
    # Required: Postgres commands like CREATE DATABASE cannot run in a transaction
    connection.autocommit = True 
    print("Connection to PostgreSQL server successful")

    cursor = connection.cursor()

    # Define and execute the statement
    create_db_query = sql.SQL("CREATE DATABASE {}").format(sql.Identifier(new_dbname))
    cursor.execute(create_db_query)
    
    print(f"Database '{new_dbname}' created successfully")

    cursor.close()
    connection.close()

except Exception as error:
    print(f"Error creating the database: {error}")
```

# Create a Table

```python
import psycopg2

# Database connection parameters
host = "localhost"
port = "5432"
dbname = "cos101_db"
user = "postgres"
password = "cos101"

try:
    # Connect to the PostgreSQL database
    connection = psycopg2.connect(
        host=host,
        port=port,
        dbname=dbname,
        user=user,
        password=password
    )
    #connection.autocommit = True  # Enable autocommit mode
    
    print("Connection to PostgreSQL DB successful")

    # Create a cursor object
    cursor = connection.cursor()

    # Define the SQL statement to create a new table
    create_table_query = """
    CREATE TABLE sst_info (
        id SERIAL PRIMARY KEY,
        name VARCHAR(50) NOT NULL,
        age INT NOT NULL,
        department VARCHAR NOT NULL,
        password VARCHAR NOT NULL
    );
    """

    # Execute the SQL statement
    cursor.execute(create_table_query)
    print("Table created successfully")

    # Close the cursor and connection
    cursor.close()
    connection.close()

except Exception as error:
    print(f"Error creating table: {error}")
```

# INSERT into Table

```python
import psycopg2

# Database connection parameters
host = "localhost"
port = "5432"
dbname = "cos101_db"
user = "postgres"
password = "cos101"

try:
    # 1. Connect to the PostgreSQL database
    connection = psycopg2.connect(
        host=host,
        port=port,
        dbname=dbname,
        user=user,
        password=password
    )
    # Enable autocommit so changes are saved immediately
    connection.autocommit = True
    print("Connection to PostgreSQL DB successful")

    cursor = connection.cursor()

    # 2. Create the table if it doesn't exist (Fixes the 'relation does not exist' error)
    create_table_query = """
    CREATE TABLE IF NOT EXISTS sst_info (
        id SERIAL PRIMARY KEY,
        name VARCHAR(50) NOT NULL,
        age INT NOT NULL,
        department VARCHAR NOT NULL,
        password VARCHAR NOT NULL
    );
    """
    cursor.execute(create_table_query)
    print("Table check complete (created or already exists)")

    # 3. Define the SQL INSERT statement 
    # We omit 'id' so the SERIAL type generates it automatically
    insert_query = """
    INSERT INTO sst_info (name, age, department, password)
    VALUES (%s, %s, %s, %s);
    """
    
    # Data to be inserted (4 items to match the 4 %s placeholders)
    data_to_insert = ("Jamel Ojo", 20, "Software Engineering", "anthelo4")

    # 4. Execute the SQL statement
    cursor.execute(insert_query, data_to_insert)
    print("Data inserted successfully")

    # 5. Clean up
    cursor.close()
    connection.close()
    print("Connection closed")

except Exception as error:
    print(f"Error: {error}")
```

```python
import psycopg2

# Fix 1: Use proper dictionary syntax
db_params = {
    "host": "localhost",
    "port": "5432",
    "dbname": "cos101_db",
    "user": "postgres",
    "password": "cos101"
}

connection = None
cursor = None

try:
    connection = psycopg2.connect(**db_params)
    cursor = connection.cursor()

    # Fix 2: Add '*' to select all columns
    select_query = "SELECT * FROM sst_info WHERE age < 30"

    cursor.execute(select_query)
    results = cursor.fetchall()

    # Fix 3: Match the variable name 'results'
    if not results:
        print("No records found matching the criteria.")
    else:
        for row in results:
            print(row)

except Exception as error:
    print(f"Error: {error}")

finally:
    if cursor:
        cursor.close()
    if connection:
        connection.close()
```
