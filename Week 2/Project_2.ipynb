```python
import pandas as pd
import numpy as np

def solve_with_pandas():
    print("Advanced Equation Solver (Pandas Integrated)")
    print("1. Quadratic | 2. Cubic | 3. Quartic")
    
    choice = input("\nSelect equation type (1-3): ")
    
    # Define labels based on the degree of the equation
    if choice == '1':
        labels = ['A (x²)', 'B (x)', 'C (constant)']
    elif choice == '2':
        labels = ['A (x³)', 'B (x²)', 'C (x)', 'D (constant)']
    elif choice == '3':
        labels = ['A (x⁴)', 'B (x³)', 'C (x²)', 'D (x)', 'E (constant)']
    else:
        print("Invalid Selection.")
        return

    # Collect inputs into a list
    coeff_values = [ ]
    for label in labels:
        val = float(input(f"Enter coefficient {label}: "))
        coeff_values.append(val)

    # Use Pandas Series to store and manage the coefficients
    coeff_series = pd.Series(coeff_values, index=[l.split(' ')[0] for l in labels])
    
    print("\n--- Input Coefficients ---")
    print(coeff_series)

    # Calculate roots using numpy's solver
    raw_roots = np.roots(coeff_series.values)

    # Use a Pandas DataFrame to display the results cleanly
    roots_df = pd.DataFrame({
        'Root Number': [f"Root {i+1}" for i in range(len(raw_roots))],
        'Value': raw_roots
    })

    print("\n--- Calculated Roots ---")
    # Setting the Root Number as index for a cleaner table look
    print(roots_df.set_index('Root Number'))

if __name__ == "__main__":
    solve_with_pandas()
```
