ALGORITHM FOR THE PROJECT_4
START
INPUT: Ask the user for the staff's Years of Experience and Age.
CHECK condition 1: If Experience > 25 AND Age >= 55:
      ATR = ₦5,600,000.
CHECK Condition 2: Else if Experience > 20 AND Age >=45:
      ATR = ₦4,480,000.
CHECK Condition 3: Else if Experience > 10 AND Age >= 35:
      ATR = ₦1,500,000.
DEFAULT: If none of the above are true (specifically for < 10 years experience and < 35 years age):
      ATR = ₦550,000.
OUTPUT: Display the calculated ATR to the user.
END

```python
def calculate_atr():
    print(" Izifin Technology: Staff ATR Calculator ")
    
    # Getting inputs from the user
    try:
        experience = float(input("Enter years of work experience: "))
        age = int(input("Enter age of the staff: "))
        
        # Applying the conditional logic provided in the brief
        if experience > 25 and age >= 55:
            atr = 5600000
        elif experience > 20 and age >= 45:
            atr = 4480000
        elif experience > 10 and age >= 35:
            atr = 1500000
        else:
            # Covers staff < 10 years experience and < 35 years age
            atr = 550000
            
        # Displaying the result with currency formatting
        print(f"\nThe Annual Tax Revenue (ATR) for this staff is: ₦{atr:,}")
        
    except ValueError:
        print("Invalid input. Please enter numbers only for age and experience.")

if __name__ == "__main__":
    calculate_atr()
```
