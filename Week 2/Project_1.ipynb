```python
BEGIN

Initialize TotalCost to 0.

LOOP while there are more items to scan:

Scan item.

Get item price.

Add item price to TotalCost.

Display TotalCost.

Prompt customer for PaymentAmount.

IF PaymentAmount is less than TotalCost:

Request more funds or cancel transaction.

ELSE:

Calculate Change = PaymentAmount - TotalCost.

Display Change.

Print Receipt.

END
```

```python
QUESTION 3: DEVELOP AN ALGORITHM THAT TAKES AS INPUT TWO NAMES AND THEIR CORRESPONDING AGE AND THEN SWAP THE AGES
```

```python
START

INPUT: Get the first person's name (Name1) and age (Age1).

INPUT: Get the second person's name (Name2) and age (Age2).

CREATE a temporary storage space called TempAge.

PROCESS: Copy the value of Age1 into TempAge. (Age1 is now safe in the temporary spot).

PROCESS: Copy the value of Age2 into Age1. (Age1 now holds the second person's age).

PROCESS: Copy the value of TempAge into Age2. (Age2 now holds the first person's original age).

OUTPUT: Display Name1 with the new Age1 and Name2 with the new Age2.

END
```
