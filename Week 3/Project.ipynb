```python
import tkinter as tk
from tkinter import messagebox
import pandas as pd

def verify():
    # 1. Get user input and make it lowercase to avoid errors
    name = name_entry.get().strip().lower()
    dept = dept_entry.get().strip().lower()

    try:
        # 2. Read the employee file
        data = pd.read_csv("GIG-logistics.csv")

        # 3. Create lowercase versions of the data for easy checking
        # This ignores if the user types 'logistics' or 'Logistics'
        names_lower = data['Name'].str.lower()
        depts_lower = data['Department'].str.lower()

        # 4. Check if the input exists in the data
        if ((names_lower == name) & (depts_lower == dept)).any():
            # Find everyone in that department
            team = data[data['Department'].str.lower() == dept]['Name'].tolist()
            
            # Remove the current user's name from the team list
            if name_entry.get().strip() in team:
                team.remove(name_entry.get().strip())

            # Show welcome message
            team_str = "\n".join(team)
            messagebox.showinfo("Success", f"Welcome!\n\nYour Team Members:\n{team_str}")
        else:
            # If not found
            messagebox.showwarning("Error", "Employee does not exist.")

    except Exception:
        messagebox.showerror("Error", "Check if GIG-logistics.csv is in the folder.")

# --- GUI SETUP ---
root = tk.Tk()
root.title("GIG Verification")

tk.Label(root, text="Name:").pack()
name_entry = tk.Entry(root)
name_entry.pack()

tk.Label(root, text="Department:").pack()
dept_entry = tk.Entry(root)
dept_entry.pack()

tk.Button(root, text="Login", command=verify).pack()

root.mainloop()
```
