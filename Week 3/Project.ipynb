```python
import pandas as pd
import tkinter as tk
from tkinter import messagebox

# Load the CSV file

data = pd.read_csv("GIG-logistics(in).csv")

# Function to check employee
def check_employee():
    name = name_entry.get().strip()
    dept = dept_entry.get().strip()

    # Clear previous result
    result_text.delete(1.0, tk.END)

    # Check if inputs are empty
    if not name or not dept:
        messagebox.showwarning("Input Error", "Please enter both name and department.")
        return

    # Search in dataset (case insensitive)
    result = data[
        (data['Name'].str.lower() == name.lower()) &
        (data['Department'].str.lower() == dept.lower())
    ]

    if not result.empty:
        result_text.insert(tk.END, f"Welcome {name}!\n\n")
        result_text.insert(tk.END, f"Other members in {dept} department:\n")

        # Get other employees in same department
        others = data[
            (data['Department'].str.lower() == dept.lower()) &
            (data['Name'].str.lower() != name.lower())
        ]

        if not others.empty:
            for person in others['Name']:
                result_text.insert(tk.END, f"- {person}\n")
        else:
            result_text.insert(tk.END, "No other members found.\n")
    else:
        result_text.insert(tk.END, "Sorry, this employee does not exist.")

# Create GUI window
root = tk.Tk()
root.title("GIG Employee Verification")
root.geometry("400x400")

# Labels and Entry fields
tk.Label(root, text="Enter Name:").pack(pady=5)
name_entry = tk.Entry(root, width=30)
name_entry.pack(pady=5)

tk.Label(root, text="Enter Department:").pack(pady=5)
dept_entry = tk.Entry(root, width=30)2
dept_entry.pack(pady=5)

# Button
tk.Button(root, text="Verify Employee", command=check_employee).pack(pady=10)

# Text box for results
result_text = tk.Text(root, height=12, width=45)
result_text.pack(pady=10)
# Run the app
root.mainloop()
```
